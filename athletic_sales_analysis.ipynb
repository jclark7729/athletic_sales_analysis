```python
# Read the CSV files into DataFrames.

df_2020 = pd.read_csv("athletic_sales_2020.csv")
df_2021 = pd.read_csv("athletic_sales_2021.csv")
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.
dtypes = df_2020.dtypes
print(dtypes)
```

```python
# Check the 2021 sales data types.
dtypes = df_2021.dtypes
print(dtypes)
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
combined_df = pd.concat([df_2020, df_2021], axis=0)
combined_df = combined_df.reset_index(drop=True)
```

```python
# Check if any values are null.
null_counts = combined_df.isnull().sum()
print(null_counts)
```

```python
# Check the data type of each column
dtypes = combined_df.dtypes
print(dtypes)
```

```python
# Convert the "invoice_date" to a datetime datatype
combined_df['invoice_date'] = pd.to_datetime(combined_df['invoice_date'], infer_datetime_format=True)
```

```python
# Confirm that the "invoice_date" data type has been changed.
dtypes = combined_df.dtypes
print(dtypes)
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
count_nonzero = combined_df.groupby(['region', 'state', 'city'])['units_sold'].sum()
print(count_nonzero)

# Rename the sum to "Total_Products_Sold".
sum_column = count_nonzero.rename('Total_Products_Sold')

# Show the top 5 results.
top_5_results = sum_column.head(5)
display(top_5_results)
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.

pivot_table = pd.pivot_table(combined_df, values='units_sold', index=['region', 'state', 'city'], aggfunc='sum')

# Rename the "units_sold" column to "Total_Products_Sold"
units_sold_column = pivot_table.rename(columns={'units_sold': 'Total_Products_Sold'})

# Show the top 5 results.
top_5_results = units_sold_column.head(5)
display(top_5_results)
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.
groupbype_df = combined_df.groupby(['region', 'state', 'city'])
total_sales = groupbype_df['total_sales'].sum()

# Rename the "total_sales" column to "Total Sales"
total_sales = groupbype_df['total_sales'].sum().rename('Total Sales')

# Show the top 5 results.
top_5_results = total_sales.head(5)
display(top_5_results)
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.
pivot_table = pd.pivot_table(combined_df, values='total_sales', index=['region', 'state', 'city'])



# Optional: Rename the "total_sales" column to "Total Sales"
pivot_table = pivot_table.rename(columns={'total_sales': 'Total Sales'})

# Show the top 5 results.
top_5_results = pivot_table.head(5)
display(top_5_results)
```

### 4. Determine which Retailer had the Most Sales

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
groupbype_df = combined_df.groupby(['retailer', 'region', 'state', 'city'])
total_sales = groupbype_df['total_sales'].sum()
# Rename the "total_sales" column to "Total Sales"
groupbype_df = groupbype_df['total_sales'].sum().rename('Total Sales')

# Show the top 5 results.
top_5_results = groupbype_df.head(5)
display(top_5_results)
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.
filtered_df = combined_df[combined_df['product'] == 'Women\'s Athletic Footwear']
display(filtered_df)
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
groupbype_df = filtered_df.groupby(['retailer', 'region', 'state', 'city'])
total_sales = groupbype_df['units_sold'].sum()

# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
groupbype_df = groupbype_df['units_sold'].sum().rename('Womens_Footwear_Units_Sold')

# Show the top 5 results.
top_5_results = groupbype_df.head(5)
display(top_5_results)
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
pivot_table = pd.pivot_table(filtered_df, values='units_sold', index=['retailer', 'region', 'state', 'city'])

# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
pivot_table = pivot_table.rename(columns={'units_sold': 'Womens_Footwear_Units_Sold'})

# Show the top 5 results.
top_5_results = pivot_table.head(5)
display(top_5_results)
```

### 6. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
pivot_table = pd.pivot_table(filtered_df, values='total_sales', index='invoice_date')

# Optional: Rename the "total_sales" column to "Total Sales"
pivot_table = pivot_table.rename(columns={'total_sales': 'Total Sales'})

# Show the table.
display(pivot_table)
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
pivot_table = pivot_table.resample('D').sum()

# Sort the resampled pivot table in descending order on "Total Sales".
pivot_table = pivot_table.sort_values(by='Total Sales', ascending=False)

display(pivot_table)
```

### 7.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
pivot_table = pivot_table.resample('W').sum()

# Sort the resampled pivot table in descending order on "Total Sales".
pivot_table = pivot_table.sort_values(by='Total Sales', ascending=False)

display(pivot_table)
```
